```python
import requests
from bs4 import BeautifulSoup
from itertools import zip_longest
import csv
import re
from pandas import read_csv
from ar_corrector.corrector import Corrector
import numpy as np
import pandas as pd
```

```python
Author = []
title = []
links = []
desc = []
category = []
page_num = 1

numbers = []
no_of_pages = []


while page_num < 200:
    # target URL
    url = f"https://www.arab-books.com//page/{page_num}"


    result = requests.get(url)

    src = result.content
    #print(src)


    # parse content
    soup = BeautifulSoup(src, 'lxml')
    #print(soup)


    Authors = soup.find_all("div", {"class":"book-writer"})
    #print(Authors)

    titles = soup.find_all("div", {"class":"excerpt-book"})
    #print(titles)



    for i in range(len(Authors)):
        Author.append(Authors[i].text.strip().replace('الكاتب',''))
        title.append(titles[i].text.strip())
        links.append(titles[i].find("a").attrs["href"])

    Author = [s.replace("\n", "") for s in Author]

    title = [s.replace("PDF", "") for s in title]
    title = [s.replace("\n", "") for s in title]

    #print(links)
    #print(Author)
    #print(title)
    page_num +=1
    if page_num == 110:
        page_num+=1
    print("page switched", page_num)



for link in links:
    result = requests.get(link)
    src = result.content
    soup = BeautifulSoup(src,"lxml")

    #department = soup.find_all("span", {"class": "post-cat-wrap"})
    department = soup.find_all("div", {"class": "entry-header"})


    descrip = soup.find("div", {"class":"book-info"}).ul
    desc_text = ""
    for li in descrip.find_all("li"):
        desc_text += li.text + "/ "
    desc_text = desc_text[:-2]
    book = {}
    numbers.append(re.findall(r'\d{1,3}',desc_text))
    infos = list(soup.find("div", class_='book-info').stripped_strings)
    for info in infos:
        if info == 'لغة الكتاب:':
            book["language"] = infos[infos.index(info)+1]
        if info == 'عدد الصّفحات:':
            book["pages"] = infos[infos.index(info)+1]
        if info == 'دار النشر:':
            book["publishing house"] = infos[infos.index(info)+1]
        if info == 'حجم الكتاب:':
            book["size"] = infos[infos.index(info)+1]
        if info == 'ملف الكتاب:':
            book["format"] = infos[infos.index(info)+1]
        if info == 'قسم الكتاب:':
            book["category"] = infos[infos.index(info)+1][6:]
    category.append(book["category"].replace(' اسلامية',''))

    
#print(numbers)
for m in numbers:
    if len(m) ==0:
        m.append(1)
        no_of_pages.append(m[0])
    else :
        no_of_pages.append(m[0])

#print(no_of_pages)

#print(len(dep))
#print(len(no_of_pages))   
file_list = [Author , title , category, links, no_of_pages]
imported = zip_longest(*file_list)

with open("D:\en.csv","w" , encoding='utf-8-sig') as project :
    wr = csv.writer(project)
    wr.writerow(["Author","Title", "Category", "Link for PDF" , "Number Of Pages" ])
    wr.writerows(imported)
```
